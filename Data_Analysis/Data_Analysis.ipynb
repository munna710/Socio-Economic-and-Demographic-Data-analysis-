```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python


# Read the dataset
df = pd.read_csv(r"C:\Users\user\OneDrive\Desktop\census_income_data2.csv")
```

```python
df.head()
```

```python
# Replace missing values in Mother_Nativity and Father_Nativity columns based on Citizenship column
mask = df['Citizenship'] == 'Native- Born in the United States'
df.loc[mask, 'Mother_Nativity'] = 'United-States'
df.loc[mask, 'Father_Nativity'] = 'United-States'
```

```python
df.head()
```

```python
# Standardize the format in the Income_Bracket column
df['Income_Bracket'] = df['Income_Bracket'].replace({'50000+.': '>50000', '-50000': '<=50000'})
```

```python
# Convert Survey_Year to four-digit year format
df['Survey_Year'] = df['Survey_Year'].apply(lambda x: 1900 + int(x))
```

```python
df.head()
```

```python
# Define the condition
condition = (df['Personal_Nativity'].isnull()) & (df['Mother_Nativity'].isnull()) & (df['Father_Nativity'].isnull()) & (df['Citizenship'] == 'Foreign born- Not a citizen of U S')

# Update the values
# If Personal_Nativity, Mother_Nativity, and Father_Nativity are null
# and Citizenship is 'Foreign born- Not a citizen of U S', then change the values to 'Foreign'
df.loc[condition, ['Personal_Nativity', 'Mother_Nativity', 'Father_Nativity']] = 'Foreign'
```

```python
# Calculate summary statistics
summary_stats = df[['Dividends', 'Capital_Gains', 'Capital_Losses']].describe()
print(summary_stats)

# Plot histograms
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['Dividends'], kde=True)
plt.title('Dividends Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['Capital_Gains'], kde=True)
plt.title('Capital Gains Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Capital_Losses'], kde=True)
plt.title('Capital Losses Distribution')

plt.tight_layout()
plt.show()

# Plot box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Dividends'])
plt.title('Dividends Box Plot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['Capital_Gains'])
plt.title('Capital Gains Box Plot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Capital_Losses'])
plt.title('Capital Losses Box Plot')

plt.tight_layout()
plt.show()
```

```python
# Handling Outliers
# Trim extreme values (e.g., top 1% and bottom 1%)
trim_low = df[['Dividends', 'Capital_Gains', 'Capital_Losses']].quantile(0.01)
trim_high = df[['Dividends', 'Capital_Gains', 'Capital_Losses']].quantile(0.99)
df_trimmed = df[(df[['Dividends', 'Capital_Gains', 'Capital_Losses']] >= trim_low) & (df[['Dividends', 'Capital_Gains', 'Capital_Losses']] <= trim_high)]

# Summary Statistics
summary_stats = df_trimmed[['Dividends', 'Capital_Gains', 'Capital_Losses']].describe()
print(summary_stats)

# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df_trimmed['Dividends'], kde=True)
plt.title('Dividends Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df_trimmed['Capital_Gains'], kde=True)
plt.title('Capital Gains Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df_trimmed['Capital_Losses'], kde=True)
plt.title('Capital Losses Distribution')

plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=df_trimmed['Dividends'])
plt.title('Dividends Box Plot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df_trimmed['Capital_Gains'])
plt.title('Capital Gains Box Plot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df_trimmed['Capital_Losses'])
plt.title('Capital Losses Box Plot')

plt.tight_layout()
plt.show()
```

```python
# Check for missing values and replace them with appropriate values
df.fillna(0, inplace=True)  # Replace NaN with 0

# Calculate Hourly_Pay only for rows where Hourly_Pay is currently zero
condition = (df['Hourly_Pay'] == 0)

# Check for division by zero in Weeks_Worked
zero_weeks_worked = (df['Weeks_Worked'] == 0)

# Calculate Hourly_Pay only for rows where Hourly_Pay is currently zero and Weeks_Worked is not zero
hourly_pay_calculation = (df['Dividends'] - (df['Capital_Gains'] + df['Capital_Losses'])) / df['Weeks_Worked']
default_hourly_pay = 10  # Set a default value for Hourly_Pay when numerator is zero
df.loc[condition & ~zero_weeks_worked, 'Hourly_Pay'] = hourly_pay_calculation
df.loc[condition & (hourly_pay_calculation == 0), 'Hourly_Pay'] = default_hourly_pay

# Handle cases where Weeks_Worked is zero (avoid division by zero)
df.loc[zero_weeks_worked, 'Hourly_Pay'] = 0

# Calculate average hourly pay for positive values
average_hourly_pay = df[df['Hourly_Pay'] > 0]['Hourly_Pay'].mean()

# Replace negative hourly pay values with average hourly pay for positive values
df.loc[df['Hourly_Pay'] < 0, 'Hourly_Pay'] = average_hourly_pay
```

```python
# Demographic Profiles
def plot_demographics(df):
    demographic_cols = ['Age', 'Race', 'Gender', 'Ancestry']
    for col in demographic_cols:
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()

# Employment Patterns
def plot_employment_patterns(df):
    employment_cols = ['Worker_Class', 'Industry', 'Occupation']
    for col in employment_cols:
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()

# Educational Attainment
def plot_education_attainment(df):
    df.groupby('Education_Level')['Income_Bracket'].value_counts().unstack().plot(kind='bar', stacked=True)
    plt.title('Income Bracket by Education Level')
    plt.xlabel('Education Level')
    plt.ylabel('Count')
    plt.show()

# Income Analysis
def income_analysis(df):
    df['Net_Capital'] = df['Capital_Gains'] - df['Capital_Losses']
    df.groupby('Income_Bracket')[['Net_Capital', 'Dividends']].mean().plot(kind='bar')
    plt.title('Average Capital Gains, Losses, and Dividends by Income Bracket')
    plt.xlabel('Income Bracket')
    plt.ylabel('Average Amount')
    plt.show()
    

# Work-Life Balance
def work_life_balance(df):
    df.plot(kind='scatter', x='Weeks_Worked', y='Hourly_Pay')
    plt.title('Correlation between Weeks Worked and Hourly Pay')
    plt.xlabel('Weeks Worked')
    plt.ylabel('Hourly Pay')
    plt.show()

# Migration Patterns
def migration_patterns(df):
    migration_cols = ['MSA_Change', 'Region_Change', 'Within_Region_Move', 'Sunbelt_Migration']
    for col in migration_cols:
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()

# Family Dynamics
def family_dynamics(df):
    df.groupby('Marital_Status')['Dependents'].value_counts().unstack().plot(kind='bar', stacked=True)
    plt.title('Dependents by Marital Status')
    plt.xlabel('Marital Status')
    plt.ylabel('Count')
    plt.show()

# Economic Impact
def economic_impact(df):
    df.groupby('Business_Ownership')['Employers_Count'].mean().plot(kind='bar')
    plt.title('Average Number of Employers by Business Ownership')
    plt.xlabel('Business Ownership')
    plt.ylabel('Average Number of Employers')
    plt.show()

# Veteran Affairs
def veteran_affairs(df):
    df.groupby('Veteran_Benefits')['Hourly_Pay'].mean().plot(kind='bar')
    plt.title('Average Hourly Pay by Veteran Benefits')
    plt.xlabel('Veteran Benefits')
    plt.ylabel('Average Hourly Pay')
    plt.show()

# Run all analysis functions
plot_demographics(df)
plot_employment_patterns(df)
plot_education_attainment(df)
income_analysis(df)
work_life_balance(df)
migration_patterns(df)
family_dynamics(df)
economic_impact(df)
veteran_affairs(df)
```
